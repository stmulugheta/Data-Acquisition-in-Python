##Data Acquisition
#There are various formats for a dataset, .csv, .json, .xlsx  etc. The dataset can be stored in different places (local machine or sometimes online.
Here, I'm going to practice loading a dataset into the Jupyter Notebook.
I'm going to use the Automobile Dataset from an online source. This data set is in CSV format.


```python
#importing libraries
import pandas as pd
import numpy as np
```

```python
#Reading the online file by the URL, and assigning it to variable "df"
#Setting the header to none so that pandas dont assume the first row is the header
URL = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"
df = pd.read_csv(URL, header=None)
```

```python
#Checking the first 5 rows using dataframe.head() method

print("The first 5 rows of the dataframe:") 
df.head(5)
```

```python
#Checking the last 5 rows using dataframe.tail() method

print("The final 5 rows of the dataframe:")
df.tail(5)
```

#Looking at the table, we need some headers to describe each column. In the section below, I will assign the headers to the table using the information from the online source:https://archive.ics.uci.edu/ml/datasets/Automobile

```python
# creating headers list (column names in order)

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
#Now we can replace the header with the list created

df.columns= headers
print(df.head(5))
```

#Here I'm going to replace the missing values (?) with NaN (Not a Number) and then use the dropna() method to remove all the missing values in a specific column; such as in 'normalized-losses'.

```python
df1=df.replace('?', np.NaN)
df1.dropna(subset=["price"], axis=0)
```

##Saving Dataset

```python
df.to_csv("automobile.csv", index=False)
```
